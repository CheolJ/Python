```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 3, 300)
exp = np.exp(t)
fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(t, exp)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_yscale('log')
ax.plot(t, exp)
```

```python
p = np.linspace(0.001, 0.999, 300)
information = -np.log2(p)

fig, ax = plt.subplots(figsize=(7, 7))
ax.set_yscale('log', base=2)
ax.plot(p, information)
```
