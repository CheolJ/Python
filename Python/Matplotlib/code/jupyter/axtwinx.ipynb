```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
PI = np.pi
t = np.linspace(0.01, 5*PI, 100)
sin = np.sin(t)
exp = np.exp(t)

fig = plt.figure(figsize=(10, 7))
ax1 = fig.add_subplot()
ax1.plot(t, sin)
ax1.plot(t, exp)
```

```python
fig = plt.figure(figsize=(10, 7))
ax1 = fig.add_subplot()
ax1.plot(t, sin)
ax2 = ax1.twinx()
ax2.plot(t, exp)
```
