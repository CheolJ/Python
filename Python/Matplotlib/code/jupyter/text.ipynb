```python
import matplotlib.pyplot as plt
```

```python
figsize = (7, 7)
fig, ax = plt.subplots(figsize=figsize)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.grid()
ax.tick_params(axis='both', labelsize=15)
ax.text(x=0, y=0, s='hello1', fontsize=30)
ax.text(x=0.5, y=0, s='hello2', fontsize=30)
ax.text(x=0.5, y=-0.5, s='hello3', fontsize=30)
```

```python
figsize = (7, 7)
fig, ax = plt.subplots(figsize=figsize)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.grid()
ax.tick_params(axis='both', labelsize=15)
ax.text(x=0, y=0, 
        va='center', ha='left', 
        s='Hello', 
        fontsize=30)
ax.text(x=0, y=0, 
        va='center', ha='center',
        s='Hello',
        fontsize=30,
        color='darkblue',
        alpha=0.5)
ax.text(x=0, y=0, 
        va='center', ha='right',
        s='Hello',
        fontsize=30,
        color='green',
        alpha=0.5)
ax.text(x=0, y=0, 
        va='top', ha='center',
        s='Hello',
        fontsize=30,
        color='red',
        alpha=0.5)
ax.text(x=0, y=0, 
        va='bottom', ha='center',
        s='Hello',
        fontsize=30,
        color='orange',
        alpha=0.5)
```

```python
figsize = (7, 7)
fig, ax = plt.subplots(figsize=figsize)

ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])

ax.grid()
ax.tick_params(axis='both', labelsize=15)

ax.text(x=-4, y=4,
        va='top', ha='left',
        s='Hello1',
        fontsize=30)
ax.text(x=4, y=4,
        va='top', ha='right',
        s="Hello2",
        fontsize=30)
ax.text(x=-4, y=-4,
        va='bottom', ha='left',
        s="Hello3",
        fontsize=30)
ax.text(x=4, y=-4,
        va='bottom', ha='right',
        s="Hello4",
        fontsize=30)
```
